### Analysis of EARTHQUAKES in INDIA


```python
edata=pd.read_excel('India_Earthquakes.xlsx')
edata
```

```python
cat=[]

for i in edata['Dis_Mag_Value']:
    if i>=5 and i<=5.9:
        cat.append("moderate")
    elif i>=6 and i<=6.9:
        cat.append("strong")
    elif i>=7 and i<=7.9:
        cat.append("major")
    elif i>=8:
        cat.append("great")
```

```python
edata['Earthquake_Category'] = cat
```

```python
edata
```

### Number of earthquakes in each year

```python
from collections import Counter
import matplotlib.pyplot as plt
```

```python
count = Counter(edata['Year'])
```

```python
keys = count.keys()
values = count.values()

plt.figure(figsize=(20,5))
plt.bar(keys, values)
plt.xlabel("Year")
plt.ylabel("No. of Earthquakes")
plt.title("No. of Earthquakes in each year")
```

### CPI of each earthquake

```python
plt.figure(figsize=(10,9))
plt.plot(edata['CPI'],edata['Earthquake_Name'], color = 'red', marker = 'o')
plt.grid(True)
plt.xlabel("CPI Number")
plt.ylabel("Earthquake_Name")
plt.title("CPI of each earthquake")
```

### Earthquake Category

```python
countCat = Counter(edata['Earthquake_Category'])
```

```python
key = []
value = []

for x, y in countCat.items():
    key.append(x)
    value.append(y)
```

```python
plt.figure(figsize=(5,5))
plt.pie(value,labels=key,autopct='%1.1f%%', radius = 3.0)
plt.title('Earthquake_Category')
plt.axis('equal')
plt.show()
```

### Total Damages caused due to earthquakes in USD	

```python
plt.figure(figsize=(10,9))
plt.plot(edata['Total_Damages_USD'],edata['Earthquake_Name'], color = 'green', marker = 'o')
plt.grid(True)
plt.xlabel("Total_Damages_USD")
plt.ylabel("Earthquake_Name")
plt.title("Total Damages caused due to earthquakes in USD")
```

### Total deaths in each category

```python
dict = {
    'great':0,
    'moderate':0,
    'strong':0,
    'major':0
}

for i in range(0,26):
    if edata['Earthquake_Category'][i] == 'great':
        dict['great'] = dict['great'] + edata['Total_Deaths'][i]
        
    elif edata['Earthquake_Category'][i] == 'moderate':
        dict['moderate'] = dict['moderate'] + edata['Total_Deaths'][i]
        
    elif edata['Earthquake_Category'][i] == 'strong':
        dict['strong'] = dict['strong'] + edata['Total_Deaths'][i]
        
    elif edata['Earthquake_Category'][i] == 'major':
        dict['major'] = dict['major'] + edata['Total_Deaths'][i]

    
dict    
    
```

```python
keys = dict.keys()
values = dict.values()

plt.figure(figsize=(10,20))
plt.bar(keys, values)
plt.xlabel("Earthquake_Category")
plt.ylabel("No. of Deaths")
plt.title("Total Deaths in each category")
```

```python
plt.figure(figsize=(5,5))
plt.pie(values,labels=keys,autopct='%1.1f%%', radius = 3.0)
plt.title("Total Deaths in each category")
plt.axis('equal')
plt.show()
```

### Total Affected in each category

```python
dict = {
    'great':0,
    'moderate':0,
    'strong':0,
    'major':0
}

for i in range(0,26):
    if edata['Earthquake_Category'][i] == 'great':
        dict['great'] = dict['great'] + edata['Total_Affected'][i]
        
    elif edata['Earthquake_Category'][i] == 'moderate':
        dict['moderate'] = dict['moderate'] + edata['Total_Affected'][i]
        
    elif edata['Earthquake_Category'][i] == 'strong':
        dict['strong'] = dict['strong'] + edata['Total_Affected'][i]
        
    elif edata['Earthquake_Category'][i] == 'major':
        dict['major'] = dict['major'] + edata['Total_Affected'][i]

    
dict    
    
```

```python
keys = dict.keys()
values = dict.values()
plt.figure(figsize=(10,5))
plt.pie(values,labels=keys,autopct='%1.1f%%', radius = 3.0)
plt.title("Total Affected in each category")
plt.axis('equal')
plt.show()
```
